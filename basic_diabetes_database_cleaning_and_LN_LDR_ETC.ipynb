```python
# 1. Importing the Dataset and reading the CSV

import pandas as pd
from sklearn import datasets

data = pd.read_csv('diabetes.csv')
df = pd.DataFrame(data) 


print(df)

# 2. Encoding outcome as they are not numerical

from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()


# 3. Dividing the Frame Now as DATA and OUTCOME

enc = label_encoder.fit(df['Outcome'])
df['Outcome'] = enc.transform(df['Outcome']) # 2. Encode outcome as they are not numerical
print(df)

dta = pd.DataFrame(df, columns=['Pregnancies',
                                'Glucose',
                                'BloodPressure',
                                'SkinThickness',
                                'Insulin',
                                'BMI',
                                'DiabetesPedigreeFunction',
                                'Age']) # 3. Divite the frame as data & outcome
print(dta)

outcome = df['Outcome']
print(outcome)  


# 4. Splitting 20% as testing data

from sklearn import model_selection 
train_x , x_test , train_y , y_test= model_selection.train_test_split(dta, outcome ,test_size=0.2)

# 5. Fitting Datasets for Logistic Regression, Linear Discriminant Analysis, GaussianNB, and DecisionTreeClassifier

from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score

models = []
models.append(('LR', LogisticRegression()))
models.append(('LDR', LinearDiscriminantAnalysis()))
models.append(('NB', GaussianNB()))
models.append(('CART', DecisionTreeClassifier()))


name = []
results = []
results1 = []
results2 = []
results3 = []

for n, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=7)
    cv_results = model_selection.cross_val_score(model, train_x, train_y, cv=kfold, scoring='accuracy')
    cv_results1 = model_selection.cross_val_score(model, train_x, train_y, cv=kfold, scoring='precision_weighted')
    cv_results2 = model_selection.cross_val_score(model, train_x, train_y, cv=kfold, scoring='recall_weighted')        
    cv_results3 = model_selection.cross_val_score(model, train_x, train_y, cv=kfold, scoring='f1_weighted')
    results.append(cv_results)
    results1.append(cv_results1)
    results2.append(cv_results2)
    results3.append(cv_results3)
    name.append(n)
    msg = "%s: %f (%f)" %(n, cv_results.mean(), cv_results.std())
    print(msg)

# Compare Algorithms based on Accuracy
fig = plt.figure()
fig.suptitle('Algorithm Comparison based on Accuracy')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(name)
plt.show()

# Compare Algorithms based on Precision Weight
fig = plt.figure()
fig.suptitle('Algorithm Comparison based on Precision Weight')
ax = fig.add_subplot(111)
plt.boxplot(results1)
ax.set_xticklabels(name)
plt.show()

# Compare Algorithms based on F1 Weight
fig = plt.figure()
fig.suptitle('Algorithm Comparison based on Recall Weight')
ax = fig.add_subplot(111)
plt.boxplot(results2)
ax.set_xticklabels(name)
plt.show()

# Compare Algorithms based on F1 Weight
fig = plt.figure()
fig.suptitle('Algorithm Comparison based on F1 Weight')
ax = fig.add_subplot(111)
plt.boxplot(results3)
ax.set_xticklabels(name)
plt.show()
```
